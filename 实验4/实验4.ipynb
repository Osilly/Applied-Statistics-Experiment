# T检验

## 做出假设

假设：男性大学生消费支出与女性大学生消费支出水平相近

## 处理脏数据（部分数据带有空格）

```python
import types
men = []
for x in df1['平均月支出'].values:
    if(type(x) is not int):
        x = ''.join(x.split())
    men.append(x)
# df1['平均月支出'].values = ''.join(df1['平均月支出'].values.split())
men = list(map(int, men))
men
```

```python
women = []
for x in df2['平均月支出'].values:
    if(type(x) is not int):
        x = ''.join(x.split())
    women.append(x)
women = list(map(int, women))
women
```

## 进行levene检验，检验两总体是否具有方差齐性

```python
stats.levene(men, women)
```

p值远大于0.05，认为两总体具有方差齐性。

## 进行两独立样本T检验

```python
stats.ttest_ind(men,women)
```

p值为0.18947719784107386，大于α（0.05），所以接受原假设
结论为：男性大学生消费支出与女性大学生消费支出水平相近

# Z检验

## 作出假设

与上述假设相同

## 进行两独立样本Z检验

```python
import statsmodels.stats.weightstats as sw
sw.ztest(men, women, value=0)
```

p值为0.18561069860827706，大于α（0.05），所以接受原假设
结论为：男性大学生消费支出与女性大学生消费支出水平相近
